```python
import pandas as pd

bank_data = pd.read_csv('bank-additional/bank-additional-full.csv', sep = ';')
bank_data.head()
```

```python
# Find any nulls in the data 
bank_data.isna().sum()
```

```python
bank_data.describe()
```

```python
# Cleaning the Data:

# Approach 1: Mean-Based Approach
# For this approach, we will remove any values that are more than 3 SD away from the mean 
import numpy as np
from scipy import stats

# We want to remove all rows that have outliers in at least one column: 
# Note: We want to only calculate the Z-Scores of columns that are numeric (columns above in describe())
# df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
approach_1 = bank_data[(np.abs(stats.zscore(bank_data[['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']])) < 3).all(axis = 1)]
print("Originally, our dataset had", len(bank_data), "rows. Now, it has", len(approach_1), "rows. This means we identified", (len(bank_data) - len(approach_1)), "rows as outliers in this approach!")
approach_1
```

```python
# Approach 2: IQR Approach
# For this approach, we will be calculating the Inter Quartile Ranges of the data
# We will need to define a function to do this
def iqr_outliers(df):
    # Lets create a set that contains all the indexes to be removed
    indexes = set()
    for col in ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']:
        Q1 = np.percentile(df[col], 25, interpolation = 'midpoint')
        Q3 = np.percentile(df[col], 75, interpolation = 'midpoint')
        IQR = Q3 - Q1
        upper = Q3 + (1.5 * IQR)
        lower = Q1 - (1.5 * IQR)
        new_outliers = df.loc[(df[col] < lower) | (df[col] > upper)]
        for idx in list(new_outliers.index):
            indexes.add(idx)
    return(indexes)


outlier_idxs = iqr_outliers(bank_data)
approach_2 = bank_data.copy()
approach_2.drop(outlier_idxs, axis = 0, inplace = True)
print("Originally, our dataset had", len(bank_data), "rows. Now, it has", len(approach_2), "rows. This means we identified", (len(bank_data) - len(approach_2)), "rows as outliers in this approach!")
approach_2
```
